```python
train_df = pd.read_csv(os.path.join(DATA_PATH, 'train.csv'))
test_df = pd.read_csv(os.path.join(DATA_PATH, 'test.csv'))
valid_df = pd.read_csv(os.path.join(DATA_PATH, 'valid.csv'))

ehr_database = joblib.load(os.path.join(DATA_PATH, 'ehr_preprocessed_seq_by_day_cat_embedding.pkl'))
```

```python
def generate_dataset(index_df:pd.DataFrame, database: dict, multiple=False, return_y=True):
    """generate data from database(EHR DICT) according to the index_df given

    Args:
        index_df (pd.DataFrame): the subjeects index dataframe
        database (dict): dictionary of EHR where the feature_dict comprises of all the ehr data of each subjects
        multiple (bool, optional): True for all stays as X, False for the last stay as X
        return_y (bool, optional): return y for train, valid subsets. Defaults to True.
    """
    X, y = [], []
    for idx, row in index_df.iterrows():
        s_idx = row['id']
        if multiple:
            X.append(database['feat_dict'][s_idx])
        else:
            X.append(database['feat_dict'][s_idx][-1])
        if return_y:
            y.append(int(row['readmitted_within_30days']))
    return X, y if return_y else X

train_x, train_y = generate_dataset(train_df, ehr_database, multiple=False)
test_x = generate_dataset(test_df, ehr_database, multiple=False, return_y=False)
val_x, val_y = generate_dataset(valid_df, ehr_database, multiple=False)

```

```python
train_data_df = pd.DataFrame(np.concatenate([train_x, np.array(train_y).reshape(-1, 1)], axis=1), columns=features_cols + ['readmitted_within_30days'])
valid_data_df = pd.DataFrame(np.concatenate([val_x, np.array(val_y).reshape(-1, 1)], axis=1), columns=features_cols + ['readmitted_within_30days'])
test_data_df = pd.DataFrame(test_x, columns=features_cols)
```

```python
# the standard deviation of features
std_train = np.std(train_x, axis=0)
std_train_dict = dict(zip(features_cols, std_train))
pd.DataFrame([std_train_dict])
```

```python
# those features with 0 standard deviation are useless as they are the same for all the patients
feature_cols_new = [key for key, value in std_train_dict.items() if value] + ['readmitted_within_30days']
len(feature_cols_new)
```

```python
corr_train = x.corr()["readmitted_within_30days"][:-1]
```

```python
print("feature_corr_positive max: ", feature_cols_new[corr_train.argmax()])
print("feature_corr_negative_max: ", feature_cols_new[corr_train.argmin()])
print("feature_corr_inpact max:", feature_cols_new[corr_train.abs().argmax()])
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(train_data_df[feature_cols_new[:-1]].values, train_data_df[feature_cols_new[-1]].values)
```

```python
y_pred = lr_model.predict(valid_data_df[feature_cols_new[:-1]].values)
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
print("Linear Regression accuracy: ", accuracy_score(valid_data_df[feature_cols_new[-1]].values, y_pred))
print("Linear Regression f1-score: ", f1_score(valid_data_df[feature_cols_new[-1]].values, y_pred))
print("Linear Regression aucroc: ", roc_auc_score(valid_data_df[feature_cols_new[-1]].values, y_pred))
```
